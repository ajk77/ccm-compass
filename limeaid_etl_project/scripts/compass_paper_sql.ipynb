```python
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect('../data/db/mydb.sqlite')
cursor = conn.cursor()

# Query the distinct count of visit_occurrence_id in VISIT_DETAIL
cursor.execute("SELECT COUNT(DISTINCT visit_occurrence_id) FROM VISIT_DETAIL")
result = cursor.fetchone()

print(f"Distinct count of visit_occurrence_ids: {result[0]}")

# Close the connection
conn.close()
```

```python
# HOSPITALIZATION_COMPASS #
# Connect to the SQLite database
conn = sqlite3.connect('../data/db/mydb.sqlite')
cursor = conn.cursor()

# Query the distinct count of visit_occurrence_id in VISIT_DETAIL
cursor.execute("SELECT age_at_admit, COUNT(age_at_admit) FROM HOSPITALIZATION_COMPASS GROUP BY age_at_admit")
result = cursor.fetchall()
print(f"Age at admit: {result}")

# Close the connection
conn.close()


# Expand counts into individual observations
ages = []
for age, count in result:
    ages.extend([age] * count)

# Plot histogram
plt.figure()
plt.hist(ages, bins=range(min(ages), max(ages) + 2))
plt.xlabel("Age at Hospital Admission")
plt.ylabel("Number of Admissions")
plt.title("Distribution of Age at Hospital Admission")
plt.tight_layout()
plt.savefig('../data/figs/age_at_admit.png')
plt.show()
```

```python
# HOSPITALIZATION_COMPASS #
# Connect to the SQLite database
conn = sqlite3.connect('../data/db/mydb.sqlite')
cursor = conn.cursor()

# Query the distinct count of visit_occurrence_id in VISIT_DETAIL
cursor.execute("SELECT location_id, COUNT(location_id) FROM HOSPITALIZATION_COMPASS GROUP BY location_id")
result = cursor.fetchall()
print(f"Location: {result}")

# Close the connection
conn.close()

location_counts = result
# Mapping of location codes to hospital names
location_map = {
    100: "UPMC Altoona",
    102: "UPMC Chautauqua",
    105: "UPMC East",
    106: "UPMC Hamot",
    109: "UPMC Jameson",
    111: "UPMC Magee - Womens Hospital",
    113: "UPMC Mercy",
    115: "UPMC Northwest",
    116: "UPMC Passavant - McCandless",
    117: "UPMC Presbyterian",
    118: "UPMC Shadyside",
    119: "UPMC St. Margaret",
    122: "UPMC Williamsport",
}

# Separate values for plotting
codes = [c for c, _ in location_counts]
counts = [n for _, n in location_counts]
labels = [location_map[c] for c in codes]

# Plot (categorical histogram via bar chart)
plt.figure()
plt.bar(labels, counts)
plt.xlabel("Hospital")
plt.ylabel("Number of Admissions")
plt.title("Admissions by Hospital Location")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig('../data/figs/location_admits.png')
plt.show()
```

```python
# HOSPITALIZATION_COMPASS #
# Connect to the SQLite database
conn = sqlite3.connect('../data/db/mydb.sqlite')
cursor = conn.cursor()

# Query the distinct count of visit_occurrence_id in VISIT_DETAIL
cursor.execute("SELECT icu_los_hours, COUNT(icu_los_hours) FROM HOSPITALIZATION_COMPASS GROUP BY icu_los_hours")
result = cursor.fetchall()
print(f"icu_los: {result}")

# Close the connection
conn.close()

# Expand counts into individual observations
loss = []
for los, count in result:
    if los is None: continue
    los = int(los // 24)
    if los > 40: continue
    loss.extend([los] * count)

# Plot histogram
plt.figure()
plt.hist(loss, bins=range(min(loss), max(loss) + 2))
plt.xlabel("ICU Length of Stay")
plt.ylabel("Number of Admissions")
plt.title("Distribution of ICU Length of Stay by Hospital Admission")
plt.tight_layout()
plt.savefig('../data/figs/ICU_los.png')
plt.show()
```
